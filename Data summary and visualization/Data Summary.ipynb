```python
import pandas as pd
import numpy as np

xlfile = 'db_score.xlsx'

#엑셀 파일 읽기
df = pd.read_excel(xlfile)

midterm = df['midterm']
final = df['final']
score = df['score']

midterm_mean = midterm.mean() #midterm 평균값 구하기
midterm_median = midterm.median() #midterm 중간값 구하기

final_mean = final.mean() #final 평균값 구하기
final_median = final.median() #final 중간값 구하기

score_mean = score.mean() #score 평균값 구하기
score_median = score.median() #score 중간값 구하기

#소수점 둘째 자리까지 나타냄
print("midterm의 mean : %0.2f"%midterm_mean)
print("midterm의 median : " + str(midterm_median))
print()
print("final의 mean : %0.2f"%final_mean)
print("final의 median : " + str(final_median))
print()
print("score의 mean :  %0.2f"%score_mean)
print("score의 median : %0.2f"%score_median)

```

```python
grade = df['grade']

grade_mode = grade.mode()#grade의 mode 구하기
#print(grade_mode) 
#A,B,C,D 모두 최빈값인데 같은 값이라서 4개가 나옴.
print( "grade의 mode : " + str(grade_mode[0]) +","+ str(grade_mode[1]) + ","+ str(grade_mode[2]) + ","+ str(grade_mode[3]) )
```

```python
midterm_var = midterm.var() #midterm의 분산값 구하기
midterm_std = midterm.std() #midterm의 표준편차 구하기

final_var = final.var()  #final의 분산값 구하기
final_std = final.std()  #final의 표준편차 구하기

score_var = score.var() #score의 분산값 구하기
score_std = score.std() #score의 표준편차 구하기

#소수점 둘째 자리까지 나타냄
print("midterm의 variance : %0.2f"%midterm_var)
print("midterm의 standard deviation: %0.2f"%midterm_std)
print()
print("final의 variance :  %0.2f"%final_var)
print("final의 standard deviation: %0.2f"%final_std)
print()
print("score의 variance :  %0.2f"%score_var)
print("score의 standard deviation: %0.2f"%score_std)

```

```python
import matplotlib.pyplot as plt

x_data = np.linspace(0,100,11) #0에서 100까지 동일한 간격으로 나눈 11개의 원소를 가지는 배열 생성 [0, 10, 20, ... , 80, 90, 100]
midterm_percentile = np.percentile(midterm, x_data) #[0, 10, 20, ... , 80, 90, 100]의 백분위에 해당하는 midterm값 배열 생성

#그래프 그리기
plt.title('midterm') 
plt.xticks(x_data)
plt.plot(x_data, midterm_percentile, 'b')
plt.plot(x_data, midterm_percentile, 'b.')
plt.show()
```

```python
final_percentile = np.percentile(final, x_data) #[0, 10, 20, ... , 80, 90, 100]의 백분위에 해당하는 final값 배열 생성


#그래프 그리기
plt.title('final')
plt.plot(x_data, final_percentile, 'b')
plt.plot(x_data, final_percentile, 'b.')
plt.show()
```

```python
score_percentile = np.percentile(score, x_data)  #[0, 10, 20, ... , 80, 90, 100]의 백분위에 해당하는 score값 배열 생성

plt.title('score')
plt.xticks(x_data)
plt.plot(x_data, score_percentile, 'b')
plt.plot(x_data, score_percentile, 'b.')
plt.show()
```
