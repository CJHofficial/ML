```
from konlpy.tag import Mecab
import numpy as np
from gensim.models import Word2Vec, KeyedVectors

ko_model = Word2Vec.load("/gdrive/MyDrive/colab/data/ko.bin")

#print(len(ko_model.wv.vocab))

#print(ko_model.wv['한국'])
KOR = ko_model.wv['한국']
SEO = ko_model.wv['서울']
PAR = ko_model.wv['파리']

temp = KOR - SEO + PAR
tempList = ko_model.wv.similar_by_vector(temp)
print("● 한국 - 서울 + 파리")
print()
print("1. " + tempList[0][0] + " 2. " + tempList[1][0] + " 3. " + tempList[2][0])
print()

CS = ko_model.wv['컴퓨터'] #컴퓨터공학 없음
NS = ko_model.wv['자연'] #자연과학도 없음
PS = ko_model.wv['인문학']

temp = CS - NS + PS
tempList = ko_model.wv.similar_by_vector(temp)
print("● 컴퓨터 - 자연 + 인문학")
print()
print("1. " + tempList[0][0] + " 2. " + tempList[1][0] + " 3. " + tempList[2][0])
print()

LOVE = ko_model.wv['사랑']
BROKE = ko_model.wv['이별']

temp = LOVE + BROKE
tempList = ko_model.wv.similar_by_vector(temp)
print("● 사랑 + 이별")
print()
print("1. " + tempList[0][0] + " 2. " + tempList[1][0] + " 3. " + tempList[2][0])
print()

CCB = ko_model.wv['충청북도']
CJ = ko_model.wv['청주']
GWANG = ko_model.wv['강원도']

temp = CCB - CJ + GWANG
tempList = ko_model.wv.similar_by_vector(temp)
print("● 충청북도-청주+강원도")
print()
print("1. " + tempList[0][0] + " 2. " + tempList[1][0] + " 3. " + tempList[2][0])
print()

HUMAN = ko_model.wv['인간']
#LOVE = ko_model.wv['사랑']

temp = HUMAN - LOVE
tempList = ko_model.wv.similar_by_vector(temp)
print("● 인간 - 사랑")
print()
print("1. " + tempList[0][0] + " 2. " + tempList[1][0] + " 3. " + tempList[2][0])
print()

ART = ko_model.wv['미술']
PAINT = ko_model.wv['물감']
MUSIC = ko_model.wv['음악']

temp = ART - PAINT + MUSIC
tempList = ko_model.wv.similar_by_vector(temp)
print("● 미술 - 물감 + 음악")
print()
print("1. " + tempList[0][0] + " 2. " + tempList[1][0] + " 3. " + tempList[2][0])
print()
```
